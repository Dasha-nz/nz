```python
# Установите количество строк
N = 10  # Замените 10 на желаемое количество строк

# Генерация случайных имен
names = [f'Name_{i}' for i in range(1, N + 1)]

# Генерация случайных годов рождения в диапазоне [1965, 1990]
birth_years = np.random.randint(1965, 1991, size=N)

# Генерация годов трудоустройства в диапазоне [BirthYear + 18, 2008]
employ_years = [(birth_year + 18) for birth_year in birth_years]

# Генерация случайных зарплат в диапазоне [10000, 30000]
salaries = np.random.randint(10000, 30001, size=N)

# Создание DataFrame
data = {
    'Name': names,
    'BirthYear': birth_years,
    'EmployYear': employ_years,
    'Salary': salaries
}

df = pd.DataFrame(data)
df.head(10)
```

```python
# зарпата больше 15
df[df['Salary'] > 15000]
```

```python
# сколько месяцев всего работает сотрудник
df['Month'] = (2024 - df['EmployYear']) * 12
df.head(5)
```

```python
# добавление нового столбца, в котором умножается количество месяцев, зарплата и процент
df['Tax'] = (df['Salary'] / 100 * 13) * df ['Month']
df.head(30)
```
