```python
# чтение файла и просмотр загруженных данных
df = pd.read_csv('gdrive/My Drive/L.2.csv')
df = pd.read_excel('gdrive/My Drive/L.2.xlsx')
df.head()
```

```python
# удаление столбцов
df = df.drop(['TY'], axis=1)
df = df.drop(['JO'], axis=1)
df = df.drop(['VL'], axis=1)
df = df.drop(['SP'], axis=1)
df = df.drop(['DA'], axis=1)
df = df.drop(['SN'], axis=1)
df = df.drop(['UR'], axis=1)
df = df.drop(['EP'], axis=1)
df = df.drop(['IS'], axis=1)
df = df.drop(['T2'], axis=1)
df = df.drop(['We'], axis=1)
df = df.drop(['In'], axis=1)
df.head(10)
```

```python
# в столбце AU меняем имя и фамилию местами
def swap_words(text):

    if isinstance(text, str):
        words = text.split()
        if len(words) > 1:
            words = words[::-1]
        return ' '.join(words)
    else:
        return text
df['AU'] = df['AU'].apply(swap_words)
df.head(10)
```

```python
# заполнение пропущенный строк предыдущими значениями
df[['T1', 'AU', 'PY', 'DO', 'KW', 'AB']] = df[['T1',  'AU', 'PY', 'DO','KW', 'AB']].ffill()
df.head(50)
```

```python
# переименовала столбцы
df.rename(columns = {'T1':'Title',
                     'AU':'Author',
                     'PY':'Year', 'DO':'DOI',
                     'KW':'Key Words',
                     'AB':'Abstract'},
          inplace = True)
df.head()
```

```python
# удалила значения в строках, которые были посередине и не являлись данными
words_to_remove = ["T1","['AU', 'AU']", "PY", "['KW', 'KW']", "DO", "AB"]


pattern = '|'.join(words_to_remove)


for column in ['Title', 'DOI', 'Abstract', 'Author', 'Key Words']:
    df[column] = df[column].str.replace(pattern, '', regex=True).str.strip()

df.head()
```

```python
# удалила запятые в столбце 'Author'
df['Author'] = df['Author'].str.replace(',', '', regex=False)

df.head(15)
```

```python
#Объединенила в список
df = df.groupby(['Title', 'Year', 'DOI', 'Abstract'], as_index=True).agg({'Author': lambda x: list(x), 'Key Words': lambda x: list(x)})
df.head(10)
```

```python
# переделала индексы
df = df.reset_index()
df.head(15)
```

```python
# удалила первую строчку
df = df.drop(index=0).reset_index(drop=True)
df.head(15)
```

```python
#Поменяла местами столбцы
df = df[['Title', 'Author', 'Year', 'Key Words', 'DOI', 'Abstract']]
df.head(15)
```
